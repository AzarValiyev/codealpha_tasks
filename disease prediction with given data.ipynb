# Load and Prepare the Dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv("C:/Users/hicra/OneDrive/Desktop/kaggleData/Training.csv")

# Display basic info
print("Shape:", df.shape)
print("Columns:", df.columns)
print(df.head())
```

#  Split Features and Labels

```python
# Separate features (symptoms) and target (prognosis)
X = df.drop(columns=['prognosis'])
y = df['prognosis']

# Encode the target labels (diseases)
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

# Train the Model

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
```

# Evaluate the Model

```python
from sklearn.metrics import accuracy_score, classification_report

# Predict on test set
y_pred = clf.predict(X_test)

# Accuracy and classification report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))
```

# Predict a Disease from New Symptoms

```python
import numpy as np
import pandas as pd

# Show all symptom columns
print("Symptoms:", list(X.columns))

# Create an empty input (all symptoms = 0)
input_symptoms = np.zeros(X.shape[1])

# Set specific symptoms to 1
input_symptoms[X.columns.get_loc('itching')] = 1
input_symptoms[X.columns.get_loc('skin_rash')] = 1

# Wrap in a DataFrame with correct feature names
input_df = pd.DataFrame([input_symptoms], columns=X.columns)

# Predict disease
predicted_label = clf.predict(input_df)
predicted_disease = le.inverse_transform(predicted_label)

print("Predicted Disease:", predicted_disease[0])
```
